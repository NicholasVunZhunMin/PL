```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1jR3qRQr2ZvWYKNuv8wen_-eTZWdc5a-LLvH7iymn2zw/edit?usp=sharing')
```

```python
# 從 gsheets 的 All-whiteboard-device 載入 sheets
sheets = gsheets.worksheet('工作表1').get_all_values()
# 將 sheets1 資料載入 pd 的 DataFrame 進行分析
df = pd.DataFrame(sheets[1:], columns=sheets[0])
# 取得最前面的5筆資料
df.head()
```

```python
import datetime
```

```python
import datetime
import pytz

taiwan_tz = pytz.timezone('Asia/Taipei')
now = datetime.datetime.now(taiwan_tz)

date_str = now.strftime('%Y-%m-%d')
time_str = now.strftime('%H:%M')

print(f"自動填入日期: {date_str}")
print(f"自動填入時間: {time_str}")

item = input("請輸入品項: ")
amount = float(input("請輸入金額: "))
category = input("請輸入分類: ")
remark = input("請輸入備註: ")
```

```python
df['金額'] = pd.to_numeric(df['金額'], errors='coerce')
print(df.info())
```

```python
# 創建一個新的 DataFrame 來代表你要新增的資料
new_data = pd.DataFrame([
    {
    '日期': date_str,
    '時間': time_str,
    '品項': item,
    '金額': amount,
    '分類': category,
    '備註': remark
}])

# 使用 concat() 將新資料合併到舊的 df 中
df = pd.concat([df, new_data], ignore_index=True)
```
